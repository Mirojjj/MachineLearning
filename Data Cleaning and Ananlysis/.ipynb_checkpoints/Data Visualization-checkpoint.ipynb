```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Sales Data.csv",  encoding= 'unicode_escape')
df
```

```python
df.shape
```

```python
df.columns.values
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df.drop_duplicates()
df
```

```python
df.shape
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.drop(["Status", "unnamed1"], axis=1)
df
```

```python
pd.isnull(df).sum()  #counts the values of nulls in each column
```

```python
df.dropna(inplace=True)
df
```

```python
df
```

```python
df["Amount"].dtypes
```

```python
df["Amount"] = df["Amount"].astype('int')
```

```python
df["Amount"].dtypes
```

```python
df.describe()
```

```python
#Exploratory analysis
ax = sns.countplot(x = "Gender", data=df)
for bars in ax.containers:
    ax.bar_label(bars);
```

```python
sales_gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
plt.figure(figsize=(20, 12))
ax = sns.countplot(x = "Occupation", data=df)
for bars in ax.containers:
    ax.bar_label(bars, fontsize=12);
plt.show()
```

```python
plt.figure(figsize=(20, 12))
ax = sns.countplot(x = "State", data=df)
for bars in ax.containers:
    ax.bar_label(bars, fontsize=12);
plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
plt.figure(figsize=(22, 12))
plt.title("Highest buyers in Tihar by Occupation")
sales_age = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Occupation',y= 'Amount' ,data = sales_age)
```
